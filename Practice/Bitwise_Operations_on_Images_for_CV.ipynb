```python
img1 = cv2.imread('image1.jpg')  # or .png
img2 = cv2.imread('image2.jpg')
```

```python
# Resize to same size for bitwise operations
img1 = cv2.resize(img1, (300, 300))
img2 = cv2.resize(img2, (300, 300))
```

```python
# Convert to grayscale (optional for clarity)
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
# Display grayscale images side by side
plt.figure(figsize=(10, 4))

# Show gray1
plt.subplot(1, 2, 1)
plt.imshow(gray1, cmap='gray')
plt.title('Grayscale Image 1')
plt.axis('off')

# Show gray2
plt.subplot(1, 2, 2)
plt.imshow(gray2, cmap='gray')
plt.title('Grayscale Image 2')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Apply Bitwise Operations
bit_or = cv2.bitwise_or(gray1, gray2)
bit_and = cv2.bitwise_and(gray1, gray2)
bit_not = cv2.bitwise_not(gray1, gray2)
bit_xor = cv2.bitwise_xor(gray1, gray2)
```

```python
plt.figure(figsize=(12, 6))

# Bitwise OR
plt.subplot(2, 2, 1)
plt.imshow(bit_or, cmap='gray')
plt.title('Bitwise OR')
plt.axis('off')

# Bitwise AND
plt.subplot(2, 2, 2)
plt.imshow(bit_and, cmap='gray')
plt.title('Bitwise AND')
plt.axis('off')

# Bitwise NOT (on Image 1)
plt.subplot(2, 2, 3)
plt.imshow(bit_not, cmap='gray')
plt.title('Bitwise NOT (Image 1)')
plt.axis('off')

# Bitwise XOR
plt.subplot(2, 2, 4)
plt.imshow(bit_xor, cmap='gray')
plt.title('Bitwise XOR')
plt.axis('off')

plt.tight_layout()
plt.show()
```
